## These abbreviations represent various pollutants and air quality indicators. Here's a brief explanation of each:

## AQI (Air Quality Index): An index used to communicate how polluted the air currently is or how polluted it is forecast to become. It is often used to convey the potential health effects of air pollution.

## CO (Carbon Monoxide): A colorless, odorless gas that can be harmful when inhaled in large amounts. It is produced by the incomplete combustion of carbon-containing fuels.

## NO2 (Nitrogen Dioxide): A reddish-brown gas that is a byproduct of combustion processes, such as those occurring in vehicles and power plants. It can irritate the lungs and contribute to respiratory problems.

## O3 (Ozone): Ozone in the Earth's atmosphere is beneficial and protects us from the sun's ultraviolet rays. However, ground-level ozone, formed by the reaction of pollutants emitted by vehicles, power plants, and other sources, can be harmful to human health.

## PM10 (Particulate Matter 10): Particulate matter with a diameter of 10 micrometers or smaller. It can include dust, dirt, soot, and liquid droplets. PM10 can be inhaled into the respiratory system and may cause health problems.

## PM2.5 (Particulate Matter 2.5): Fine inhalable particles with a diameter of 2.5 micrometers or smaller. Like PM10, PM2.5 can have adverse health effects when inhaled.

## SO2 (Sulphur Dioxide): A colorless gas with a pungent odor. It is produced by the burning of fossil fuels containing sulfur, such as coal and oil. SO2 can irritate the respiratory system and contribute to air pollution.

## TS (Total Suspended Particles): This term generally refers to the total mass of airborne particles, including both PM10 and PM2.5. It provides a measure of overall particulate pollution.

## Monitoring these pollutants is crucial for assessing and managing air quality, as they can have significant impacts on human health and the environment. Governments and environmental agencies often use monitoring stations and sensors to measure these parameters and calculate the AQI.
